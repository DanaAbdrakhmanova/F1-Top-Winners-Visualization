```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
results = pd.read_csv("results.csv")
drivers = pd.read_csv("drivers.csv")

# Объединение таблиц по driverId
merged = pd.merge(results, drivers, on='driverId', how='left')

# Создание полного имени
merged['fullname'] = merged['forename'] + ' ' + merged['surname']

# Приведение позиции к числу
merged['position'] = pd.to_numeric(merged['position'], errors='coerce')

# Фильтр победителей и построение графика
top_winners = merged[merged['position'] == 1]['fullname'].value_counts().head(10)

top_winners.plot(kind='barh', title='Лучшие гонщики 1950-2024', color='red')
plt.xlabel('Количество побед')
plt.ylabel(' ')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display

df = pd.read_csv("results.csv")

# Выпадающий список гонщиков
drivers = df['driverId'].unique()
dropdown = widgets.Dropdown(options=drivers, description="Гонщик")

# Функция, которая будет вызываться при выборе
def update_graph(driver_id):
    data = df[df['driverId'] == driver_id]
    fig = px.histogram(data, x='position')
    fig.show()

widgets.interact(update_graph, driver_id=dropdown)
```

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
import ipywidgets as widgets
from IPython.display import display

# Настройка рендера
pio.renderers.default = 'notebook_connected'

# Загрузи данные
df = pd.read_csv("results.csv")  # если ещё не загружено

# Виджет
drivers = df['driverId'].unique()
dropdown = widgets.Dropdown(options=drivers, description='Гонщик:')

# Функция
def update_graph(driver):
    filtered = df[df['driverId'] == driver]
    if filtered.empty:
        print("Нет данных для выбранного гонщика")
        return
    fig = px.bar(
        filtered,
        x='raceId',
        y='positionOrder',
        title=f'Позиции гонщика {driver} по гонкам',
        labels={'raceId': 'Гонка', 'positionOrder': 'Позиция'}
    )
    fig.update_yaxes(autorange='reversed')
    fig.show()

# Отображаем виджет
display(dropdown)
widgets.interact(update_graph, driver=dropdown)
```
